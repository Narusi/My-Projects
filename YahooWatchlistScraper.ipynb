```
import requests
from bs4 import BeautifulSoup
```

```
wPath = 'https://finance.yahoo.com/u/motif/watchlists/bullet-proof-balance-sheets'
page = requests.get(wPath)
soup = BeautifulSoup(page.content, 'html.parser')
```

```
for a in soup.find_all('a'):
    print(a.text)
    print(a.get('href'))
    print(a.get('title'))
```
